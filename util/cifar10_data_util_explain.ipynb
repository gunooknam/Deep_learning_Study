```python
from __future__ import print_function  # python 3.x version
import pickle                          
import numpy as np
import os
from scipy.misc import imread
```

## with 구문을 이용한 file open => close를 안해도 된다.
pickle.load(f, encoding='latin1') 로 open하고 리턴 받은 것을 확인하면 dicionary 형태임을 확인할 수 있다.

```python
def load_CIFAR_batch(filename):
  """ load single batch of cifar """
  with open(filename, 'rb') as f:
    datadict = pickle.load(f, encoding='latin1')
    X = datadict['data']
    Y = datadict['labels']
    X = X.reshape(10000, 3, 32, 32).transpose(0,2,3,1).astype("float")
    #print(X.shape) # (10000, 32, 32, 3)
    Y = np.array(Y)
    return X, Y
```

```python
# ROOT : Cifar10의 데이터가 담긴 디렉토리
# 폴더의 내용물
# data_batch1
# data_batch2
# data_batch3 
# data_batch4
# data_batch5
# test_batch
def load_CIFAR10(ROOT):
  """ load all of cifar """
  xs = []
  ys = []
  for b in range(1,6):
    f = os.path.join(ROOT,'data_batch_%d' % (b, ))
    X, Y = load_CIFAR_batch(f)
    xs.append(X)
    ys.append(Y)
  # 10000개씩 뽑아내고 그것을 append
  
  Xtr = np.concatenate(xs)
  Ytr = np.concatenate(ys)
  del X, Y
  Xte, Yte = load_CIFAR_batch(os.path.join(ROOT, 'test_batch'))
  return Xtr, Ytr, Xte, Yte
```

# CiFAR data load example
* 위에서 정의한 함수를 불러온다.
* 타입을 확인하였더니 numpy.ndarray

```python
X_train, y_train, X_test, y_test = load_CIFAR10("./cifar-10-batches-py")

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
import matplotlib.pyplot as plt
categories = [
              "airplane",
              "automobile",
              "bird",
              "cat",
              "deer",
              "dog",
              "frog",
              "horse",
              "ship",
              "truck"
              ]

ntrain_loaded = X_train.shape[0]
batch_size = 10
                            # 총 50000개의 train set에서 Shuffle하고 
                            # batch_size만큼의 10개를 뽑느다.
randidx = np.random.randint(ntrain_loaded, size=batch_size)

# Shuffle 된 index 10개에서 loop가 돈다.


for i in randidx:
    currimg = X_train[i, :]
    currlabel= y_train[i]
    plt.figure(figsize=(1,1))
    plt.imshow(currimg.astype('uint8'))
    title_string = "[%d] %d-class %s" % (i, currlabel, categories[currlabel])
    plt.title(title_string) 
    plt.show()
```
