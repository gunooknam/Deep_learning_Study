# Python Opencv Tutorial 

* 가장 기본적인 이미지 불러오기

 * imread() 이미지를 불러올 때 사용 <p>
 * imshow() 로 이미지 display창을 띄울 수 있으나 waitKey() 함수를 같이 써야 한다.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
# 가볍게 Opencv version을 확인해주자
print("OpenCV-Python Version %s" % cv2.__version__)
```

```python
img = cv2.imread('./data/lena.PNG',cv2.IMREAD_COLOR)
# img.imshow('레나',img)

if img is None:
    raise RuntimeError('불러올 이미지가 없습니다..')
else:
    print('이미지 로드')
    print(img.shape) # 칼라이미지, 3채널 이라는 것을 알 수 있다.
    print(type(img))
    print(img.dtype)

```

```python
img2 = img[:,:,::-1]

plt.imshow(img2) # jupyter에서는 이 방법을 사용할 것이다.
```

```python
img3=cv2.resize(img2, (40,40))
print(img3.shape)
plt.figure(figsize=(1,1))
plt.imshow(img3) # jupyter에서는 이 방법을 사용할 것이다.
plt.title("this is small lena")
plt.draw()
```

```python
plt.subplot(131)
plt.imshow(img2[...,0])

plt.subplot(132)
plt.imshow(img2[...,1])

plt.subplot(133)
plt.imshow(img2[...,2])
```
