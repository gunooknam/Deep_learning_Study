```python
# torch.utils.data.Dataset 을 상속 받아서 클래스를 구성한다.
class MyCustomImageDataset(Dataset):
    def read_data_set(self):
        all_img_files = [] # => 리스트 형태로 저장
        all_labels = []    #
        class_names = os.walk(self.dataset_path).__next__()[1]
        
        # enumerate ! 
        for index, class_name in enumerate(class_names):
            label = index
            img_dir = os.path.join(self.dataset_path, class_name) # 디렉토리 넣고
            img_files = os.walk(img_dir).__next__()[2] # 디렉토리를 넣으면 파일이름은 세번쨰로 빠진다.
            # 세번째 엘리먼트는 파일이름들이다.
            for img_file in img_files:
                img_file = os.path.join(img_dir, img_file)
                img = Image.open(img_file)
                if img is not None:
                    all_img_files.append(img_file)
                    all_labels.append(label)
            
        return all_img_files, all_labels, len(all_img_files), len(class_names)
        
        
    def __init__(self, dataset_path, transforms=None):
        # dataset_path는 train 혹은 testset의 path이다.
        self.dataset_path = dataset_path
        # 여기서 주의할 것은 list형태로 넣어줘야 한다는 것
        self.image_files_path, self.labels, self.length, self.num_classes= self.read_data_set()
        # transforms 을 넣어주는 부분
        self.transforms=transforms
        
    def __getitem__(self, index): # 하나의 인덱스를 꺼낸다.
        image = Image.open(self.image_files_path[index])
        image = image.convert("RGB")
        
        if self.transforms is not None:
            image = self.transforms(image)
            
        return {'image':image,'label':self.labels[index]}
    
    def __len__(self):
        return self.length
```

```python
class CustomConvNet(nn.Module):
    
    def __init__(self, num_classes):
        super(CustomConvNet, self).__init__()    
        self.layer1 = self.conv_module(3,16)        
        self.layer2 = self.conv_module(16,32)
        self.layer3 = self.conv_module(32,64)
        self.layer4 = self.conv_module(64,128)
        self.layer5 = self.conv_module(128,256)
        self.gap = self.global_avg_pool(256,num_classes)
    
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = self.layer5(out)
        out = self.gap(out)
        out = out.view(-1, num_classes)
        return out
        
    def conv_module(self, in_num, out_num):
        return nn.Sequential(
                nn.Conv2d(in_num, out_num, kernel_size=3, stride=1, padding=1),
                nn.BatchNorm2d(out_num),
                nn.LeakyReLU(),
                nn.MaxPool2d(kernel_size=2, stride=2))
    
    def global_avg_pool(self, in_num, out_num):
        return nn.Sequential(
            nn.Conv2d(in_num, out_num, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(out_num),
            nn.LeakyReLU(),
            nn.AdaptiveAvgPool2d((1,1))) # MAP이다. 여기서 평균을 낸다.
```

```python
EPOCH = 5
BATCH_SIZE = 8
LEARNING_RATE = 0.001

transforms_train = transforms.Compose([transforms.Resize((128,128)),
                                      transforms.RandomRotation(10.),
                                      transforms.ToTensor()])

transforms_test = transforms.Compose([transforms.Resize((128,128)),
                                    transforms.ToTensor()])

train_dataset = MyCustomImageDataset(
                dataset_path="./data/Dataset/train",
                transforms=transforms_train)

trainloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)


test_dataset = MyCustomImageDataset(
                dataset_path="./data/Dataset/test",
                transforms=transforms_test)

testloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=True)

```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
num_classes = train_dataset.num_classes
print(train_dataset.num_classes)
custom_model = CustomConvNet(num_classes=num_classes).to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(custom_model.parameters(),  # 모델 파라미터를 넣고, learning_rate를 정해준다. 
                             lr = LEARNING_RATE)
```

```python
# Loss and optimizer
for e in range(EPOCH):
    for i_batch, item in enumerate(trainloader):
        images = item['image'].to(device) # GPU로 돌리자 @@
        labels = item['label'].to(device) # GPU로 돌리자 @@
        
        # Forward Pass 구간을 정하고 
        outputs = custom_model(images)
        loss = criterion(outputs, labels) #criterion 안에다가 output을 넣어서 -> labels로 만들어준다. 
        
        # Backward Pass 구간을 정한다. 
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if (i_batch + 1) % BATCH_SIZE == 0:
            print('Epoch [{}/{}], Loss: {:.4f}'
                  .format(e + 1, EPOCH, loss.item()))
            
# 모델의 웨이트를 저장한다.  
torch.save(custom_model.state_dict(),'./Customcnn.pth')
```

```python
def imshow(img):
    np_img = img.to("cpu").numpy() # 일단은 GPU 텐서를 CPU로 바꿔서 numpy(GPU가 지원이 안되므로)를 실행시킨다.
    plt.imshow(np.transpose(np_img, (1,2,0)))
    print(np_img.shape)


# Test the model 
# 웨이트를 로드한다.
custom_model.load_state_dict(torch.load('./Customcnn.pth'))
custom_model.eval()

# __getitem__만 지정해주면 알아서 배치돌아가게 나눠주는 건가?
with torch.no_grad():
    correct = 0
    total = 0
    for item in testloader:
        images = item['image'].to(device)
        labels = item['label'].to(device)
        outputs = custom_model(images)
        imshow(torchvision.utils.make_grid(images, nrow=4)) # 한번만 뜸
        _, predicted = torch.max(outputs.data, 1)
        total += len(labels)
        correct += (predicted == labels).sum().item()
        
    print('Test Accuracy of the model on the {} test images: {} %'.
         format(total, 
                100 * correct / total))
    
    
#imshow(torchvision.utils.make_grid(testloader , nrow=20//2))
#print((torchvision.utils.make_grid(testloader )).shape)
```
