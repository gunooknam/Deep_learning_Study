```python
import time
import numpy as np
import os
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
slim = tf.contrib.slim
```

```python
def safe_mkdir(path):
    """ Create a directory if there isn't one already. """
    try:
        os.mkdir(path)
    except OSError:
        pass

class LeNet(object):
    def __init__(self, learning_rate, keep_prob, n_class, n_test, batch_size ,is_training, graph_dir):
        
        self.learning_rate=learning_rate
        self.keep_prob = tf.constant(keep_prob)
        self.n_class = n_class
        self.n_test = n_test
        self.is_training = is_training
        self.batch_size = batch_size
        self.graph_dir = graph_dir
        self.gstep = tf.Variable(0, dtype=tf.int32, trainable=False, name='global_step')
        
    def get_data(self):
        with tf.name_scope('data'):
                                     # 데이터 받아오는 것은 keras거 이용
            (trd, trl), (ted, tel) = tf.keras.datasets.mnist.load_data()
            train=(trd, trl)
            test=(ted, tel)
            self.temp=test[0]
            train_data = tf.data.Dataset.from_tensor_slices(train)
            train_data = train_data.shuffle(10000) # if you want to shuffle your data
            train_data = train_data.batch(self.batch_size)
            print("Train_BatchDataset : ", train_data)
            
            
            test_data = tf.data.Dataset.from_tensor_slices(test)
            test_data = test_data.batch(self.batch_size)
            print("Test_BatchDataset : ", test_data)
            
            
            iterator = tf.data.Iterator.from_structure(train_data.output_types, 
                                                       train_data.output_shapes)
            self.img, self.label = iterator.get_next()
            self.img = tf.reshape(self.img, shape=[-1, 28, 28, 1])
            self.img = tf.cast(self.img, dtype = tf.float32)
            self.label = tf.one_hot(self.label,10) # 원핫 형태로 넣어서 cross_entropy로 비교
            
            # reshape the image to make it work with tf.nn.conv2d
            self.train_init = iterator.make_initializer(train_data)  # initializer for train_data
            self.test_init = iterator.make_initializer(test_data)    # initializer for train_data 
            
    def convNet(self):
        with tf.device("/gpu:0"):
            # Mnist라서 LeNet을 쓰지만 입력 크기는 다르다.
            # Convolutional Layer #1
            # Input Tensor Shape: [batch_size, 28, 28, 1]
            # Output Tensor Shape: [batch_size, 28, 28, 32]
            conv1 = slim.conv2d(self.img, 32, [5,5], scope='conv1') # padding=same 과 actionvation이 relu로 디폴트 고정
            #conv1 = tf.layers.conv2d( inputs=x_image, filters=32, kernel_size=[5, 5], padding="same", activation=tf.nn.relu)

            # Pooling Layer #1
            # Input Tensor Shape: [batch_size, 28, 28, 32]
            # Output Tensor Shape: [batch_size, 14, 14, 32]
            pool1 = slim.max_pool2d(conv1, [2, 2], scope='pool1')
            #pool1 = tf.layers.max_pooling2d(inputs=conv1, pool_size=[2, 2], strides=2)

            # Convolutional Layer #2
            # Input Tensor Shape: [batch_size, 14, 14, 32]
            # Output Tensor Shape: [batch_size, 14, 14, 64]
            conv2 = slim.conv2d(pool1, 64, [5,5], scope='conv2')

            # Pooling Layer #2
            # Second max pooling layer with a 2x2 filter and stride of 2
            # Input Tensor Shape: [batch_size, 14, 14, 64]
            # Output Tensor Shape: [batch_size, 7, 7, 64]
            pool2 = slim.max_pool2d(conv2,  [2,2], scope='pool2')

            # Flatten tensor into a batch of vectors
            # Input Tensor Shape: [batch_size, 7, 7, 64]
            # Output Tensor Shape: [batch_size, 7 * 7 * 64]
            pool2_flat = slim.flatten(pool2, scope='flatten')

            # 7 * 7 * 64 = 3136
            # (1 x 3136 ) (3136 X 1024 )
            # Fully connected Layer
            # Input Tensor Shape: [batch_size, 7 * 7 * 64]
            # Output Tensor Shape: [batch_size, 1024]
            fc1 = slim.fully_connected(pool2_flat, 1024, scope='fc1')
            #dense = tf.layers.dense(inputs=pool2_flat, units=1024, activation=tf.nn.relu)

            # Add dropout operation; 0.6 probability that element will be kept   #drop out은 train인 때만 한다. 
            fc1_drop = slim.dropout(fc1, keep_prob=self.keep_prob , is_training=self.is_training, scope='dropout')
            #dropout = tf.layers.dropout(inputs=dense, rate=0.4, training=is_training)

            # Logits layer
            # Input Tensor Shape: [batch_size, 1024]
            # Output Tensor Shape: [batch_size, 10]
            self.logits = slim.fully_connected(fc1_drop, 10, activation_fn=None, scope='logits')
            #return logits, is_training, x_image
        
        
    def summary(self):
        ''' 텐서보드에다가 넣을 거 작성 '''
        with tf.name_scope('summaries'):
            tf.summary.scalar('loss', self.loss)
            tf.summary.histogram('histogram_loss', self.loss)
            self.summary_op = tf.summary.merge_all()
            
            
    def loss(self):
        ''' loss function, cross_entropy 사용 '''
        with tf.name_scope('loss'):
            entropy = tf.nn.softmax_cross_entropy_with_logits(labels=self.label, logits=self.logits)
            self.loss = tf.reduce_mean(entropy, name='loss')
    
    
    def optimize(self):
        '''
        optimize function 정의
        '''
        self.opt = tf.train.AdamOptimizer(self.learning_rate).minimize(self.loss, global_step=self.gstep)

    def evaluate(self):
        with tf.name_scope('predict'):
            preds = tf.nn.softmax(self.logits)
            correct_preds = tf.equal(tf.argmax(preds, 1), tf.argmax(self.label, 1))
            self.accuracy = tf.reduce_sum(tf.cast(correct_preds, tf.float32))
        
        
    def build_network(self):
        
        self.get_data()
        self.convNet()
        self.loss()
        self.optimize()
        self.evaluate()
        self.summary()
        
        
    def train_one_epoch(self, sess, saver,init, writer, epoch, step):
        start_time = time.time()
        sess.run(init)
        
        self.training = True
        total_loss = 0
        n_batches = 0
        
        try:
            while True:
                _, l, summaries = sess.run([self.opt, self.loss, self.summary_op])
                writer.add_summary(summaries, global_step=step)
                
                if (step + 1) % 100 == 0:
                    print('Loss at step {0}: {1}'.format(step, l))
                    
                step += 1
                total_loss += l
                n_batches += 1
        except tf.errors.OutOfRangeError:
            pass
        
        saver.save(sess, 'checkpoints/convnet_layers/mnist-convnet', step)
        print('Average loss at epoch {0}: {1}'.format(epoch, total_loss/n_batches))
        print('Took: {0} seconds'.format(time.time() - start_time))
        return step
    
    
    
    def eval(self):
        '''
        Count the number of right predictions in a batch
        '''
        with tf.name_scope('predict'):
            preds = tf.nn.softmax(self.logits)
            correct_preds = tf.equal(tf.argmax(preds, 1), tf.argmax(self.label, 1))
            self.accuracy = tf.reduce_sum(tf.cast(correct_preds, tf.float32))
    
    
    
    def eval_once(self, sess, init):
        start_time = time.time()
        
        test=sess.run(init) # 이것으로 test data_set이 담긴 들어온다.
        self.training = False
        total_correct_preds = 0
        i=0
#########################################################################################################
##############  내가 test 하려고 하는데 원하는 값을 feed dict 하는 경우 feed_dict로 넣어줌 ##############
        test_batch_size = 32
        batch_index = np.random.choice(len(self.temp), size=test_batch_size, replace=False)
        batch_xs = self.temp[batch_index]
        batch_xs = np.reshape(batch_xs, (-1, 28, 28, 1))
        y_pred = sess.run(self.logits, feed_dict={self.img: batch_xs})

        fig = plt.figure(figsize=(16, 10))
        for i, (px, py) in enumerate(zip(batch_xs, y_pred)):
          p = fig.add_subplot(4, 8, i+1)
          p.set_title("y_pred: {}".format(np.argmax(py)))
          p.imshow(px.reshape(28, 28), cmap='gray')
          p.axis('off')
#########################################################################################################
        try:
            while True:
                accuracy_batch, summaries = sess.run([self.accuracy, self.summary_op])
                total_correct_preds += accuracy_batch
                
        except tf.errors.OutOfRangeError:
            pass

        print('Accuracy : {0} '.format(total_correct_preds/self.n_test))
        
        
        print('Took: {0} seconds'.format(time.time() - start_time))
    
    
    
    def test(self):
        
        sess_config = tf.ConfigProto(gpu_options=tf.GPUOptions(allow_growth=True))
        with tf.Session(config = sess_config) as sess:
            sess.run(tf.global_variables_initializer())
            saver = tf.train.Saver()
            ckpt = tf.train.get_checkpoint_state(os.path.dirname('checkpoints/convnet_layers/checkpoint'))
            if ckpt and ckpt.model_checkpoint_path:
                saver.restore(sess, ckpt.model_checkpoint_path)
                print("test start !!")
                self.eval_once(sess, self.test_init) # self.test_init 으로 test data가 담긴 데이터셋으로 측정
            else :
                print("Weight 가 없습니다. 학습을 시켜주세요.")
                
            
            
    
    def train(self, n_epochs):
        
        safe_mkdir('./checkpoints')
        safe_mkdir('./checkpoints/convnet_layers') 
        writer = tf.summary.FileWriter(self.graph_dir, tf.get_default_graph())
        
        sess_config = tf.ConfigProto(gpu_options=tf.GPUOptions(allow_growth=True))
        
        with tf.Session(config = sess_config) as sess:
            sess.run(tf.global_variables_initializer())
            saver = tf.train.Saver()
            ckpt = tf.train.get_checkpoint_state(os.path.dirname('checkpoints/convnet_layers/checkpoint'))
            # 이러한 get_checkpoint_stat 함수를 사용하면 ckpt에 mnist-convnet-7500 뽑아준다. 
            # print("os.path.dirname : ",os.path.dirname('checkpoints/convnet_layers/checkpoint'))
            # print("ckpt는: ? ", ckpt)
            # 기존의 checkpoint가 있다면 restore를 쓴다. 
            if ckpt and ckpt.model_checkpoint_path:
                 saver.restore(sess, ckpt.model_checkpoint_path)
        
            step = self.gstep.eval()
            
            for epoch in range(n_epochs):
                step = self.train_one_epoch(sess, saver, self.train_init , writer,  epoch,  step)
                #self.eval_once(sess, self.test_init, writer, epoch, step)
                
        writer.close() 
    
    
    
        
```

```python
learning_rate = 0.001
keep_prob = 0.75
n_class = 10
n_test = 10000
batch_size =16

# Train을 하려면 - True 
# Test를 하려면 - False
is_training = False
checkpoint_dir = './checkpoints/'
graph_dir = './graphs/'


if __name__ == '__main__':
    model=LeNet(learning_rate,
                keep_prob,
                n_class, 
                n_test, 
                batch_size,
                is_training,
                graph_dir)
    model.build_network( )
    if is_training:
        model.train(n_epochs=1)
    else :
        model.test()
```
